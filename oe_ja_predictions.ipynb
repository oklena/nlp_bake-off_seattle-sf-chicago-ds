```python
difference_count = 0
for word in first_document:
    p_word = p_stemmer.stem(word)
    s_word = s_stemmer.stem(word)
    
    if p_word != s_word:
        print(p_word, s_word)
        difference_count += 1
        
print("\n" f"Of the {len(first_document)} words in the sample doc, only {difference_count} are stemmed differently")
    
```
